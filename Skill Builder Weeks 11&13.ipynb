```sage
# 3.11.2 Exercise (a)

var('x, y, c')

f(x, y) = e^(x+5*y)
inner_integral = integrate(f(x, y), x, 0, 1)
outer_integral = integrate(inner_integral, y, 0, 1)

answer = outer_integral.full_simplify()
answer
```

```sage
vertices = [(0,0), (pi, 0), (0, pi)]

D = polygon(vertices, color='green', edgecolor='black')

D.show(xmin=-1, xmax=4, ymin=-1, ymax=4, aspect_ratio=1, title=r"Region $D$ bounded by $y=0, x=0, y=\pi-x$")
```

```sage
# 3.11.2 Exercise (b)
var('x, y')

# This tells the computer to assume x is less than pi so that (pi - x) is positive (otherwise the value of the integral changes)
assume(x < pi)

f(x, y) = sin(x+y)
inner_integral = integrate(f(x, y), (y, 0, pi-x))
outer_integral = integrate(inner_integral, (x, 0, pi))

answer = outer_integral.full_simplify()
answer
```

```sage
# 3.11.2 Exercise (c)
var('x, y')

f(x, y) = (x + 7*y)
inner_integral = integrate(f(x, y), (y, x^2, x))
outer_integral = integrate(inner_integral, (x, 0, 1))

answer = outer_integral.full_simplify()
answer
```

```sage
var('x y')

region = region_plot([x >= 0, x^2 + y^2 <= 4], (x, -0.5, 2.5), (y, -2.5, 2.5), incol='red', bordercol='black')

region.show(aspect_ratio=1, title=r"Region $D$: $x=0$ to $x=\sqrt{4-y^2}$")
```

```sage
# 3.11.2 Exercise (d)
var('x, y')

f(x, y) = x*y^2
inner_integral = integrate(f(x, y), (x, 0, sqrt(4-y^2)))
outer_integral = integrate(inner_integral, (y, -2, 2))

answer = outer_integral.full_simplify()
answer
```
