```python
var('x, y')
f(x) = 4*x + 1

epsilon = 1
i = 0 <  4*epsilon > abs(f(x) - 9)

P1 = plot(f(x), (x, 0, 4), color="red", legend_label=r"$f(x) = 4x + 1$", thickness=2, zorder=4)
I1 = region_plot(i, (x, 2 - epsilon, 2 + epsilon), (y, 0, 17), alpha = 0.5)
L1 = line2d(((1, 0), (1, 17)), color="gray", linestyle="--", thickness=2, legend_label=r"$|x-2| = \delta \text{ at }(x=1, 3)$", zorder=3)
L2 = line2d(((3, 0), (3, 17)), color="gray", linestyle="--", thickness=2, zorder=3)

# Outlines for the Limit
C1 = point2d((2, f(2)), marker="o", color="black", markeredgecolor="black", size=20, zorder=5)

# Edge dots to delimit the points where 4*epsilon = -(f(x) - 9) and +(f(x) - 9)
C2 = point2d((1, f(1)), marker="o", color="white", markeredgecolor="black", size=20, zorder=5)
C3 = point2d((3, f(3)), marker="o", color="white", markeredgecolor="black", size=20, zorder=5)

show_all = P1 + I1 + L1 + L2 + C1 + C2 + C3

show_all.show(aspect_ratio=0.4, title="Epsilon-Delta Visualization for f(x) = 4x + 1")
```

```python
g(x) = x^2 - 5*x + 6

c = 1
L = 2
epsilon = 1
delta = min(1, epsilon / 4)

# Plot ranges
x_min, x_max = 0, 3.5
y_min, y_max = -1, 7

# Plot the function itself
P_function = plot(g(x), (x, x_min, x_max), color="red", legend_label=r"$g(x) = x^2 - 5x + 6$", thickness=2, zorder=10)

R_epsilon = region_plot(abs(y - L) < epsilon, (x, x_min, x_max), (y, y_min, y_max), incol="blue", alpha=0.6, legend_label=r"$\epsilon$-band ($|y - 2| < 1$)")
R_delta = region_plot(abs(x - c) < delta, (x, x_min, x_max), (y, y_min, y_max), incol="purple", alpha=0.6, legend_label=r"$\delta$-interval ($|x - 1| < 0.25$)")

# Add dashed lines for c and L
L_limit = line([(x_min, L), (x_max, L)], color="blue", linestyle="--", legend_label=r"$L = 2$")
C_point = line([(c, y_min), (c, y_max)], color="purple", linestyle="--", legend_label=r"$c = 1$")

show_all = R_epsilon + R_delta + L_limit + C_point + P_function
show_all.show(xmin=x_min, xmax=x_max, ymin=y_min, ymax=y_max, aspect_ratio=0.5, title=r"Epsilon-Delta: $\lim_{x \to 1} (x^2 - 5x + 6) = 2$ with $\epsilon=1, \delta=0.25$", gridlines=True, legend_loc='upper left')
```
