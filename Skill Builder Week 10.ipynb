```python
# 3.10.2 Exercise (a)

# Defining the range of the plot
x_min, x_max = -2, 3
y_min, y_max = -2, 5

# Defining the piecewise function as f(x) and g(x)
f(x) = 2 - x^2 # for (x < 1)
g(x) = 2*x - 1 # for (x ≥ 1)

line_1 = line([(x_min, 0), (x_max, 0)], color="red", thickness=2) # y = 0
line_2 = line([(-1, y_min), (-1, y_max)], color="green", thickness=2) # x = -1
line_3 = line([(2, y_min), (2, y_max)], color="purple", thickness=2) # x = 2
p1 = plot(f(x), (x, x_min, 1), color="blue") # Plotting f(x) only on its domain (x < 1)
p2 = plot(g(x), (x, 1, x_max), color="blue") # Plotting g(x) only on its domain (x ≥ 1)

show_all = p1 + p2 + line_1 + line_2 + line_3
show_all.show()
```

```python
# 3.10.2 Exercise (b)

# Defining the range of the plot
x_min, x_max = -2, 1.5

# Defining the piecewise function as f(x) and g(x)
f(x) = cos(x) # for (-π/2 ≤ x ≤ 0)
g(x) = 1 - x # for (0 < x ≤ 1)

line_1 = line([(x_min, 0), (x_max, 0)], color="red", thickness=2) # y = 0
p1 = plot(f(x), (x, -pi/2, 0), color="blue") # Plotting f(x) only on its domain (-π/2 ≤ x ≤ 0)
p2 = plot(g(x), (x, 0, 1), color="blue") # Plotting g(x) only on its domain (0 < x ≤ 1)

show_all = p1 + p2 + line_1
show_all.show()
```
